# Damage vs Speed

```
fig = plt.figure()
spd_dmg = fig.add_subplot(1, 1, 1)
#spd_dmg.plot(totactsClean.ACCDMG2015)
spd_dmg.plot(trains.TRNSPD, trains.ACCDMG2015, '.')
spd_dmg.set_title("Accident Damage vs. Speed")
spd_dmg.set_ylabel("Cost ($10M)")
spd_dmg.set_xlabel("Speed (MPH)")
```

# Tons vs Speed

```
fig = plt.figure()
spd_dmg = fig.add_subplot(1, 1, 1)
#spd_dmg.plot(totactsClean.ACCDMG2015)
spd_dmg.plot(trains.TRNSPD, trains.TONS, '.')
spd_dmg.set_title("Train speed vs. Tonnage")
spd_dmg.set_ylabel("Tonnage")
spd_dmg.set_xlabel("Speed")
```

#Log(Damage) vs Speed

```
fig = plt.figure()
spd_dmg = fig.add_subplot(1, 1, 1)
#spd_dmg.plot(totactsClean.ACCDMG2015)
spd_dmg.plot(trains.TRNSPD, np.log(trains.ACCDMG2015+1), '.')
spd_dmg.set_title("Accident Damage vs. Tonnage")
spd_dmg.set_ylabel("Log Cost ($10M)")
spd_dmg.set_xlabel("Speed")
```

# Train Accidents 2011-2013

```
plt.figure(figsize=(8,8))
type_counts13 = trains.TYPE[trains.YEAR==13].value_counts(sort=False)
type_counts12 = trains.TYPE[trains.YEAR==12].value_counts(sort=False)
type_counts11 = trains.TYPE[trains.YEAR==11].value_counts(sort=False)
DF_type = pd.DataFrame([type_counts13,type_counts12,type_counts11],index = [2013,2012,2011])
DF_type = DF_type.transpose()
DF_type.plot(kind="bar")
```
